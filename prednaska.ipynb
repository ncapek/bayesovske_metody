## Podmíněná pravděpodobnost
**Definice (úplný systém jevů):**  
Nechť $(\Omega, A, P)$ je pravděpodobnostní prostor. Jevy $A_1, A_2, \dots \in A$ tvoří úplný systém jevů (partition), pokud $\cup^{\infty}_{k=1} A_k$ $ = \Omega$ a pro $i \neq j$ je $A_i \cap A_j = \emptyset$. Úplný systém jevů muže být i konečný.

**Definice (podmíněná pravděpodobnost):**  
Nechť $(\Omega, A, P)$ je pravděpodobnostní prostor. Podmíněná pravděpodobnost (conditional probability), jevu $C \in A$ za podmínky jevu $B \in A, P(B) >0$ je
$$P(C|B) = \frac{P(C \cap B)}{P(B)}$$

**Tvrzení (Formule úplné pravdepodobnosti /Law of total probability):**  
Nechť $A_1, A_2, \dots $ je úplný systém jevů na $(\Omega, A, P)$ se všemi $P(A_k) > 0$. Potom pro libovolný jev $B \in A$ platí
$$P(B) = \sum_{i=1}^\infty P(B|A_i)P(A_i)$$

## Bayesův vzorec pro pravděpodobnosti
**Tvrzení (Bayesův vzorec/Baye's theorem):**  
Nechť $A_1, A_2, \dots $ je úplný systém jevů na $(\Omega, A, P)$ se všemi $P(A_k) > 0$. Potom pro libovolný jev $B \in A$, $P(B) > 0$ platí
$$P(A_k|B) = \frac{P(B|A_k)P(A_k)}{\sum_{i=1}^\infty P(B|A_i)P(A_i)}, k = 1, 2, \dots $$

- $P(A_k)$ jsou apriorní pravděpodobnosti (prior), informace již známé před náhodným pokusem,<br><br>
- $P(B|A_k)$ jsou věrohodnosti (likelihood), tzn. pravděpodobnost, s jakými bychom pozorovali výsledky náhodného pokusu, kdyby daná apriorní informace byla správná,<br><br>
- $P(A_k|B)$ jsou aposteriorní pravděpodobnosti (posterior), spočítané na základě apriorních pravděpodobností a výsledku náhodného pokusu.

## Frekventistická a bayesovská statistika

## Statistická inference
Ve statistické inferenci využíváme pozorování náhodné veličiny $Y$, resp. často celého náhodného vektoru, jejíž rozdělení pravděpodobností závisí na parametru $\theta$, příp. na vektoru parametrů, k tomu, abychom určitým způsobem odhadli skutečnou hodnotu parametru $\theta$.<br>
Princip statistické inference lze rozdělit do dvou přístupů:<br>
1. klasický, tzv. frekventistický přístup
2. bayesovský přístup<br>

Klasický přístup:
- parametr $\theta$ považujeme za neznámou konstantu (příp. vektor konstant)
- k inferenci o $\theta$ využijeme rozdělení pravděpodobnosti $P(Y; \theta)$ a naměřené hodnoty $Y$.

## Bayesovský přístup
- parametr $\theta$ považujeme za náhodnou veličinu (příp. náhodný vektor) $\Theta$,
- k inferenci o $\theta$ využijeme tzv. aposteriorní rozdělení pravděpodobnosti $P(Y|\theta)$ a naměřené hodnoty $Y$,
- a dále (možná neúplnou) tzv. apriorní informaci o rozdělení pravděpodobnosti $P_\Theta$ parametru $\theta$.<br>

Apriorní informace o rozdělení pravděpodobnosti $P_\Theta$:
- objektivní - z výsledků podobných, často minulých, pozorování,
- subjektivní - založená na názorech nebo zkušenostech experta.

## Výhody a nevýhody
výhody bayesovského přístupu:
- využití apriorní informace 
- dovoluje počítat i pravděpodobnosti jevů, které nejsou výsledkem opakovaných nezávislých náhodných pokusů

nevýhody bayesovského přístupu:
- konstrukce apriorního rozdělení pravděpodobnostni parametru \theta na základě apriorní informace
- použití subjektivní apriorní informace
- uvažování parametru $\theta$ jako náhodné veličiny

Pokud je objektivní apriorní informace neznámá, subjektivní může způsobit značné problémy

## Bayesův vzorec pro absolutně spojité náhodné veličiny
Uvažme $Y, \Theta$. Předpokládejme:
- $h(\theta)$ je hustota pravděpodobnosti náhodné veličiny $\theta$
- $f(y|\theta)$ je podmíněná hustota pravděpodobnosti náhodné veličiny $Y$ za předpokladu $\Theta = \theta$

Simultánní hustota pravděpodobnosti náhodného vektoru $(\Theta, Y)$:
$$g(\theta, y) = f(y|\theta)h(\theta)$$

Marginální hustota pravděpododobnosti náhodné veličiny $Y$:
$$f(y) = \int_{-\infty}^{\infty} g(\theta, y) d\theta = \int_{-\infty}^{\infty} f(y|\theta)h(\theta) d\theta$$

Bayesův vzorec obecně popisuje podmíněnou hustotu náhodné veličiny $\Theta$ za podmínky $Y=y$:
$$h(\theta|y) = \frac{g(\theta, y)}{f(y)} = \frac{f(y|\theta)h(\theta)}{\int_{-\infty}^{\infty} f(y|\theta)h(\theta) d\theta}$$

## Model
Specifikace modelu:
- náhodný vektor $(\Theta, Y)$, kde náhodný parametr $\Theta$ i měřená veličina $Y$ jsou obvykle vícerozměrné
- parametrický prostor (parametric space) $(T, B^m, P_\Theta)$
- výběrový prostor (sample space) $(Y, B^n, P_Y)$
- apriorní hustota (prior) h(\theta): hustota náhodné veličiny $\Theta$ vzhledem k míře $v$,
- model experimentu (sampling model) $f(y|\theta)$: podmíněná hustota náhodné veličiny $Y$ vzhledem k míře $\mu$ za podmínky $\Theta = \theta$<br><br>
Výsledkem bayesovské inference je aposteriorní hustota (posterior) $h(\theta|y)$:
podmíněná hustota náhodné veličiny $\Theta$ vzhledem k míře $v$ za podmínky $Y = y$

Bayesova věta:
$$h(\theta|y) = \frac{f(y|\theta)h(\theta)}{\int_T f(y|\theta)h(\theta)dv(\theta)}$$
pokud je jmenovatel kladný, jinak $h(\theta|y) = 0$


# Přednáška 2

## Binomický model

### Odvození 1
- pozorujeme sérii n nezávislých bernoulliovských pokusů $y_1, \dots, y_n$<br><br>
$f(y_1,\dots,y_n|\theta) = \prod_{i=1}^{n} f(y_k|\theta) = \prod_{i=1}^{n} \theta^{y_k} (1-\theta)^{1-y_k} = \theta^{\sum_{k=1}^{n} y_k} (1-\theta)^{\sum_{k=1}^{n} y_k}$<br><br>
možné hodnoty pro $\theta$: $T = [0;1]$, možné hodnoty pro $y$: $Y = \{0;1\}$<br>
zvolíme apriorní rozdělení:
$$\Theta \sim Rs([0;1]), h(\theta) = 1$$
<br>
pomocné vztahy:<br>
- Gama funkce: $\Gamma(a) = \int_0^\infty x^{a-1}e^{-x}dx, a>0$<br>
- Beta funkce: $B(a, b) = \int_0^1 x^{a-1}(1-x)^{b-1}, a,b>0$
<br>
použijeme Bayesovu větu:
$$h(\theta|y_1,\dots,y_n) = \frac{f(y_1,\dots,y_n|\theta)h(\theta)}{\int_0^1f(y_1,\dots,y_n|\theta)h(\theta)d\theta} = \frac{\theta^{\sum_{k=1}^n y_k}(1-\theta)^{n-\sum_{k=1}^n y_k}1}{\int_0^1\theta^{\sum_{k=1}^n y_k}(1-\theta)^{n-\sum_{k=1}^n y_k}1d\theta} = \frac{\theta^{\sum_{k=1}^n y_k}(1-\theta)^{n-\sum_{k=1}^n y_k}}{B(1+\sum_{k=1}^n y_k, 1+n-\sum_{k=1}^n y_k)}$$<br>
tedy $\Theta(y_1,\dots,y_n) \sim Beta(a,b)$, kde $a=1+\sum y_k, b=1+n-\sum y_k$<br><br>
Postačující statistika: $Y=\sum_{k=1}^n y_k$ obsahuje veškerou informaci potřebnou pro odvození $\theta$

### Odvození 2
$Y|\Theta=\theta \sim Bi(n, \theta)$<br>
$f(y|\theta) = {n \choose y}\theta^y (1-\theta)^{n-y}$<br>
$y \in \{0,\dots,n\}$<br>
apriorní rozdělení:<br>
$\Theta \sim Rs([0;1]), h(\theta) = 1$<br><br>

$$h(\theta|y) = \frac{f(y|\theta)h(\theta)}{\int_0^1 f(y|\theta)h(\theta)d\theta} = \frac{{n \choose y} \theta^y (1-\theta)^{n-y} .1 }{\int_0^1 {n \choose y} \theta^y (1-\theta)^{n-y}} = \frac{{n \choose y} \theta^y (1-\theta)^{n-y} .1 }{{n \choose y}\int_0^1  \theta^y (1-\theta)^{n-y}}=\frac{\theta^y(1-\theta)^{n-y}}{B(1+y,1+n-y)}$$<br>
tedy aposteriorní rozdělení je:<br>
$(\Theta|y) \sim Beta(a,b)$

### Odvození 3
tentokrát apriorní rozdělení:<br>
$\Theta \sim Beta(a,b)$<br>
$h(\theta) = \frac{\theta^{a-1}(1-\theta)^{b-1}}{B(a,b)}$<br>
Experiment:
$(Y|\theta) \sim Bi(n,\theta)$<br>
$f(y|\theta) = {n \choose y}\theta^y (1-\theta)^{n-y}, Y=\{0,1,\dots,n\}$<br><br>

Aposteriorní rozdělení:<br>
$$h(\theta|y) = \frac{f(y|\theta)h(\theta)}{\int_0^1 {f(y|\theta)h(\theta)d\theta}} = \frac{{n \choose y} \theta^y (1-\theta)^{n-y} \frac{\theta^{a-1}(1-\theta)^{b-1}}{B(a,b)}}{\int_0^1 {n \choose y} \theta^n (1-\theta)^{n-y} \frac{\theta^{a-1}(1-\theta)^{b-1}}{B(a,b)}d\theta} = \frac{\theta^{y+a-1}(1-\theta)^{n-y+b-1}}{\int_0^1 a^{y+a-1} (1-\theta)^{n-y+b-1}}$$<br>
$A - 1 = y+a-1, B-1 = n-y+b-1$<br>
$A = a+y, B = b+n-y$<br>
$h(\theta|y) = \frac{\theta^{A-1} (1-\theta)^{B-1}}{B(A,B)}$<br>
$(\Theta|y) \sim Beta(A,B)$<br><br>

Apriorní:<br>
$\Theta \sim Beta(a,b)$ pro $a,b>0$<br>
$E[\Theta] = \frac{a}{a+b}$<br>
$var[\Theta] = \frac{ab}{(a+b)^2 (a+b+1)}$<br>
$Mode(\Theta) = \frac{a-1}{a+b-2}$ pro $a,b>1$<br><br>

Aposteriorni:<br>
$(\Theta|y) \sim Beta(A,B)$, kde $A=a+y, B=b+n-y$<br>
$E[\Theta|y] = \frac{A}{A+B} = \frac{a+y}{a+y+b+n-y} = \frac{a+y}{a+b+n}$<br>
$var(\Theta|y) = \frac{(a+y)(b+n-y)}{(a+b+n)^2(a+b+n+1)}$

### Interpretace paremetrů
$a$ - jakýsi fiktivní počet úspěchů ze kterého čerpáme informaci do apriorního rozdělení<br>
$b$ - stejné pro neúspěchy<br>
$a+b$ - tedy dohromady počet pokusů<br>
$E(\Theta|y) = \frac{a+y}{a+b+n} = \frac{a}{a+b+n} + \frac{y}{a+b+n} = \frac{a}{a+b+n}.\frac{a+b}{a+b} + \frac{y}{a+b+n}.\frac{n}{n} = \frac{a+b}{a+b+n}.\frac{a}{a+b} + \frac{n}{a+b+n}.\frac{y}{n}$<br>
$\frac{a+b}{a+b+n}, \frac{n}{a+b+n}$ jsou váhy, které v součtu dají 1<br>
$\frac{a}{a+b}$ je frekventistickýý odhad pravděpodobnosti úspěchu fiktivního experimentu z apriorního rozdělení<br>
$\frac{y}{n}$ je frekventistický odhad $\theta=\frac{y}{n}$<br>
tedy: $E(\Theta|y) = \frac{a+b}{a+b+n}.E(\Theta) + \frac{n}{a+b+n}\bar {y}$<br>
tedy jde o vážený průměr apriorní hodnoty $E(\Theta)$ a odhadu $\bar{y}$<br>
větší hodnoty $a,b$ přiřazují větší hodnoty apriornímu rozdělení<br>
zároveň to funguje i pro $n=0$ a $n=1$

### Predikce
- apriorní prediktivní rozdělení (prip predictive distribution) je rozdělení pravděpodobnosti náhodné veličiny $Z \in Y$ dokud nejsou pozorována žádná data. Pro apriorní prediktivní hustotu platí:<br>
$$f(z) = \int_T {g(z,\theta)}dv(\theta) = \int_T {f(z|\theta)h(\theta)dv(\theta)}$$<br><br>

- Aposteriorní prediktivní rozdělení (posterior predictive distribution) je rozdělení pravděpodobnosti podmíněné náhodné veličiny $(Z|Y_1=y_1,\dots,Y_n=y_n) \in Y$ pro pozorování dat $(y_1,\dots,y_n)$. Pro aposteriorní prediktivní hustotu platí:<br>
$$f(z|y_1,\dots,y_n) = \int_T g(z, \theta|y_1,\dots,y_n)dv(\theta) = \int_T f(z|\theta,y_1,\dots,y_n)h(\theta|y_1,\dots,y_n)dv(\theta) = \int_T f(z|\theta)h(\theta|y_1,\dots,y_n)dv(\theta)$$

#### Příklad
Modelujeme pravděpodobnost $\theta$ vyrobení zmetku v nějakém výrobním procesu, $\theta \in T = [0;1]$.<br>
Výroba součástky probíhá v továrně každý den. Nastavení strojů se každý den může mírně změnit a proto pravděpodobnost vyrobení zmetku považujeme za náhodnou veličinu s beta rozdělením:<br>
$$\Theta \sim B(a,b), a,b>0$$<br>

Apriorní hustota je tedy:<br>
$$h(\theta) = \frac{1}{B(a,b)} \theta^{a-1}(1-\theta)^{b-1}$$<br>
Náhodný pokus: po $n$ dnů sledujeme výrobu, náhodná veličina $Y$ je počet vyrobených zmetků, tzn.<br>
$$Y \sim Bi(n,\Theta), n \in N$$<br>
a model je tedy zadán podmíněnou pravděpodobnostní funkcí<br>
$$f(y|\theta) = {n \choose y} \theta^y (1-\theta)^{n-y}$$<br>
Dosadíme to Bayesovy věty a dostaneme:<br>
$$h(\theta|y) = \frac{\theta^{a+y-1}(1-\theta)^{b+n-y-1}}{B(a+y, b+n-y)}$$<br>
Tedy aposteriorní rozdělení je opět beta, ovšem s jinými parametry, které jsou ovlivněny $n$ a pozorovaným počtem zmetků $y$.<br><br>

## Princip věrohodnosti
Funkce věrohodnosti (likelihood) je systém funkcí<br>
$l_y(\theta) = \{cf(y|\theta);$ c nezávislé na $\theta\}$

## Poissonův model
$(y_1, \dots, y_n)$ náhodný výběr z $Po(\theta)$ rozdělení, $y_k$ = počty sledovaných událostí

Pravděpodobnostní funkce $Y_k|\theta$ je $f(y_k|\theta) = e^{-\theta}\frac{\theta^{y_k}}{k!}$<br>
$(Y_k|\Theta=\theta) \sim Po(\theta), f(y_1,\dots,y_n|\theta) = exp[-n\theta]\prod_{k=1}^{n} \frac{\theta^{y_k}}{y_k!}$<br>

parametrický prostor $T = [0;\infty)$, výběrový prostor: $y = \mathbb{N}_0$<br><br>

aposteriorní šance:
$$\frac{h(\theta_a | y_1,\dots,y_n)}{h(\theta_b | y_1,\dots,y_n)} = \frac{\frac{f(y_1,\dots,y_n|\theta_a)}{\int_0^{\infty} f(y_1,\dots,y_n|\theta_a)}}{\frac{f(y_1,\dots,y_n|\theta_b)}{\int_0^\infty f(y_1,\dots,y_n|\theta_b)}} = \frac{f(y_1,\dots,y_n|\theta_a)}{f(y_1,\dots,y_n|\theta_b)} = \frac{e^{-n\theta_a}\prod_{k=1}^{n} \frac{\theta_a^{y_k}}{y_k!}h(\theta_a)}{e^{-n\theta_b}\prod_{k=1}^{n} \frac{\theta_b^{y_k}}{y_k!}h(\theta_b)} = e^{-n(\theta_a-\theta_b)}(\frac{\theta_a}{\theta_b})^{\sum_{k=1}^n y_k}.\frac{h(\theta_a)}{h(\theta_b)}$$<br>
postačující statistika pro $\theta$: $Y = \sum_{k=1}^n Y_k$<br>
$(Y|\Theta=\theta) \sim Po(n\theta)$<br>
$f(y|\theta) = exp[-n\theta]\frac{(n\theta)^y}{y!}$<br><br>

Když známe postačující statistiku, můžeme odvodit aposteriorní rozdělení pomocí ni, ulehčí to výpočet:<br>
$$h(\theta|y) = \frac{f(y|\theta)h(\theta)}{\int_0^{\infty}f(y|\theta)h(\theta)d\theta} $$
